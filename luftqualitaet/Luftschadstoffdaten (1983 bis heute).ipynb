```python
import pandas as pd
```

```python
# load a file (adapt path if needed), NOTE: replace the Windows-default backslash (`\`) in the path with a forward slash (`/`)

# download the files ugz_luftqualitaetsmessung_seit-2012.csv and ugz_luftqualitaetsmessungen_1983-2011.csv from
# https://data.stadt-zuerich.ch/dataset/luftqualitaet-historisierte-messungen

# luftqualitätsmessung ostluft (2012 until now)
data = pd.read_csv('ugz_luftqualitaetsmessung_seit-2012.csv', encoding='utf-8', header=[0,1,2,3,4,5], parse_dates=[0])
data.head()
```

```python
# luftqualitätsmessung ostluft (1983-2011)
old_data = pd.read_csv('ugz_luftqualitaetsmessungen_1983-2011.csv', encoding='utf-8', header=[0,1,2,3,4,5], parse_dates=[0])
old_data.head()
```

```python
all_data = pd.concat([old_data, data])
all_data
```

```python
# get data for one station
stampf_data = all_data[['Datum', 'Zürich Stampfenbachstrasse']]
stampf_data
```

```python
# get rid of multiple levels and add column for station
stampf_data.columns = ["Datum", "Feinstaub PM10", "Globalstrahlung", "Kohlenmonoxid", "Luftdruck", "Luftfeuchte relativ", "Lufttemperatur", "Ozon, Anzahl Stundenmittel > 120 µg/m3", "Ozon, höchstes Stundenmittel", "Regendauer", "Schwefeldioxid", "Stickstoffdioxid", "Stickstoffmonoxid", "Windgeschwindigkeit skalar"]
stampf_data.set_index('Datum', inplace=True, drop=False)
stampf_data = stampf_data.assign(Station='Zürich Stampfenbachstrasse')
stampf_data
```

```python
# histograms of some attibutes
stampf_data.hist(column='Schwefeldioxid')
stampf_data.hist(column='Kohlenmonoxid')
stampf_data.hist(column='Feinstaub PM10')
stampf_data.hist(column='Lufttemperatur')
stampf_data.hist(column='Regendauer')
```

```python
# show progression of some values over time
stampf_data.plot.line(y='Lufttemperatur')
```

```python
# show progression of some values over time
stampf_data.plot.line(y='Schwefeldioxid', figsize=(20,15))
```

```python
# show last 400 entries in detail
stampf_data.tail(400).plot.line(y='Lufttemperatur')
# show last 100 entries in detail
stampf_data.tail(100).plot.line(y='Lufttemperatur')
```

```python
# compare air temperature and Ozon
stampf_data.tail(100).plot.line(y=['Lufttemperatur', 'Ozon, höchstes Stundenmittel'])
```

```python
# compare last years values with this years values
stampf_data['day_of_year'] = stampf_data['Datum'].dt.dayofyear
stampf_2012 = stampf_data.query('Datum < "2013-01-01" and Datum >= "2012-01-01"')
stampf_2013 = stampf_data.query('Datum < "2014-01-01" and Datum >= "2013-01-01"')
stampf_2014 = stampf_data.query('Datum < "2015-01-01" and Datum >= "2014-01-01"')
stampf_2015 = stampf_data.query('Datum < "2016-01-01" and Datum >= "2015-01-01"')
stampf_2016 = stampf_data.query('Datum < "2017-01-01" and Datum >= "2016-01-01"')
stampf_2017 = stampf_data.query('Datum < "2018-01-01" and Datum >= "2017-01-01"')
stampf_2018 = stampf_data.query('Datum < "2019-01-01" and Datum >= "2018-01-01"')
stampf_2019 = stampf_data.query('Datum < "2020-01-01" and Datum >= "2019-01-01"')

stampf_2012.set_index('day_of_year', inplace=True)
stampf_2013.set_index('day_of_year', inplace=True)
stampf_2014.set_index('day_of_year', inplace=True)
stampf_2015.set_index('day_of_year', inplace=True)
stampf_2016.set_index('day_of_year', inplace=True)
stampf_2017.set_index('day_of_year', inplace=True)
stampf_2018.set_index('day_of_year', inplace=True)
stampf_2019.set_index('day_of_year', inplace=True)

stampf_2012 = stampf_2012.add_suffix('_2012')
stampf_2013 = stampf_2013.add_suffix('_2013')
stampf_2014 = stampf_2014.add_suffix('_2014')
stampf_2015 = stampf_2015.add_suffix('_2015')
stampf_2016 = stampf_2016.add_suffix('_2016')
stampf_2017 = stampf_2017.add_suffix('_2017')
stampf_2018 = stampf_2018.add_suffix('_2018')
stampf_2019 = stampf_2019.add_suffix('_2019')

joined_data = stampf_2012.join(stampf_2013).join(stampf_2014).join(stampf_2015)
joined_data = joined_data.join(stampf_2016).join(stampf_2017).join(stampf_2018).join(stampf_2019)
joined_data
```

```python
joined_data.plot.line(y=['Lufttemperatur_2013', 'Lufttemperatur_2014'], figsize=(20,15))
```

```python
joined_data.plot.line(y=['Lufttemperatur_2015', 'Lufttemperatur_2016'], figsize=(20,15))
```
