```python
SSL_VERIFY = True
# evtl. SSL_VERIFY auf False setzen wenn die Verbindung zu https://www.zuerich.com nicht klappt (z.B. wegen Proxy)
# Um die SSL Verifikation auszustellen, bitte die nächste Zeile einkommentieren ("#" entfernen)
SSL_VERIFY = False
if not SSL_VERIFY:
    import urllib3
    urllib3.disable_warnings()
```

```python
# download CSV
import base64
from IPython.display import HTML

def create_download_link( df, title = "Download CSV file", filename = "data.csv"):  
    csv = df.to_csv(None, index=False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<h3><a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a><h3>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
prev_ckan_pkg_data_path = r'\\szh.loc\ssz\applikationen\OGD\Daten\CKAN-API\_Archiv\20190304_Bestand_Datasets_CKAN_PKG_LIST.xlsx'
data = pd.read_excel(prev_ckan_pkg_data_path)
data.head()
```

```python
data.columns
```

```python
prev_showcases = data[data.type == 'showcase']
prev_showcases = prev_showcases.astype({'num_datasets': 'int32'})
prev_showcases
```

```python
from ckanapi import RemoteCKAN, NotFound
site = RemoteCKAN('https://data.stadt-zuerich.ch')
```

```python
n = 500
page = 1
showcases = []
while True:
    search_params = {
        'fq': 'dataset_type:showcase',
        'rows': n,
        'start': n * (page - 1),
    }
    
    try:
        new_showcases = site.call_action('package_search', search_params, requests_kwargs={'verify': SSL_VERIFY})
        count = len(new_showcases['results'])
        if count:
            print("%s showcases added" % count)
            showcases.extend(new_showcases['results'])
            page = page + 1
        else:
            print("No more showcases.")
            break
    except NotFound:
        print("Couldn't find more packages (page %s)" % page)
```

```python
# Unfortunately we don't get all information of a showcase from the "package_search" endpoint
# so let's use "package_show" for all of them

showcases_full = []
for showcase in showcases:
    try:
        showcase_full = site.call_action('package_show', {'id': showcase['name']}, requests_kwargs={'verify': SSL_VERIFY})
        showcases_full.append(showcase_full)
        print("Got full information of showcase %s" % showcase['name'])
    except NotFound:
        print("Couldn't find showcase %s" % showcase['name'])
```

```python
new_showcases = pd.DataFrame(showcases_full)
new_showcases.head()
```

```python
all_showcases = pd.merge(prev_showcases, new_showcases, how='inner', on='id', suffixes=('_prev', '_new'))
all_showcases
```

```python
num_datasets = all_showcases[['id', 'name_new', 'num_datasets_prev', 'num_datasets_new']]
num_datasets
```

```python
diff_num_datasets = num_datasets.query("num_datasets_prev != num_datasets_new")
diff_num_datasets
```

```python
create_download_link(diff_num_datasets, filename="diff_num_datasets.csv")
```
